```python
def MJDtoDT(mjd2k):
    """Take the modified julian day from 2000, convert to seconds, add to seconds since jan 1,00:00:00.00"""
    t0 = dt.datetime.utcfromtimestamp(0.0)
    t2k = dt.datetime(2000,1,1,0,0,0)
    t2k = (t2k-t0).total_seconds()
    tmjd = mjd2k * 86400
    date = dt.datetime.utcfromtimestamp(t2k+tmjd)
    return date
```

```python
###CURVE FIT TEST
x=smallts.index
xdata = (smallts.index-t0gmt).total_seconds()
ydata = smallts['o3column']
omega_anual=2*np.pi/(365*24*60*60)
t0 = dt.datetime.fromtimestamp(0.0)
timez = timezone('Etc/GMT+6')
t0gmt = timez.localize(t0)
xp = np.average((smallts.index-t0gmt).total_seconds())

def func(x, a1,a2,a3, b, c, d, e, f, g, h, i, j):
    return (a1*pow((x-xp),1)+a2*pow((x-xp),2)+a3*pow((x-xp),3)+b) + (c*np.cos(omega_anual*x))+(d*np.sin(omega_anual*x))+(e*np.cos(2*omega_anual*x))+(f*np.sin(2*omega_anual*x))+(g*np.cos(3*omega_anual*x))+(h*np.sin(3*omega_anual*x))+(i*np.cos(4*omega_anual*x))+(j*np.sin(4*omega_anual*x))

popt, pcov = curve_fit(func, xdata, ydata)
yfitcurve = func(xdata, *popt)

f, ax = plt.subplots(figsize=(14,6))
ax.plot(x,ydata*1e18*10000/2.6867e20, color='lightblue', label = 'O3 Column')
ax.plot(x,yfitcurve*1e18*10000/2.6867e20, label = 'Fourier+Poly (Curve_Fit)')
ax.legend(fontsize=12, loc='upper left')
ax.set_xlabel('Year',size=14)
ax.set_ylabel('Dobson',size=14)
ax.set_title("Ozone profile",size=16)
```

```python
###SVD Test
#making svd matrix using a list of functions
#
#preparar la data
x=smallts.index
xdata = (smallts.index-t0gmt).total_seconds()
ydata = smallts['o3column']
omega_anual=2*np.pi/(365*24*60*60)
t0 = dt.datetime.fromtimestamp(0.0)
timez = timezone('Etc/GMT+6')
t0gmt = timez.localize(t0)
xp = np.average((smallts.index-t0gmt).total_seconds())
nl = xdata-xp

#hacer una lista de funciones de fourier para crear la matriz svd
def ff1c(x):
    return np.cos(omega_anual*x)
def ff1s(x):
    return np.sin(omega_anual*x)
def ff2c(x):
    return np.cos(2*omega_anual*x)
def ff2s(x):
    return np.sin(2*omega_anual*x)
def ff3c(x):
    return np.cos(3*omega_anual*x)
def ff3s(x):
    return np.sin(3*omega_anual*x)
def ff4c(x):
    return np.cos(4*omega_anual*x)
def ff4s(x):
    return np.sin(4*omega_anual*x)

fl = [ff1c,ff1s,ff2c,ff2s,ff3c,ff3s,ff4c,ff4s]

#hacer la matriz svd
svdmat = np.array([ydata,fl[0](xdata),fl[1](xdata),fl[2](xdata),fl[3](xdata),fl[4](xdata),fl[5](xdata),fl[6](xdata),fl[7](xdata),-pow(nl,0),-pow(nl,1),-pow(nl,2),-pow(nl,3)])

#hacer SVD y obtener el vector del minimo valor singular
U, s, Vh = linalg.svd(svdmat)
coeff = U[-1]

#rehacer la funcion
yfit_svd = coeff[1]*fl[0](xdata)+coeff[2]*fl[1](xdata)+coeff[3]*fl[2](xdata)+coeff[4]*fl[3](xdata)+coeff[5]*fl[4](xdata)+coeff[6]*fl[5](xdata)+coeff[7]*fl[6](xdata)+coeff[8]*fl[7](xdata)-coeff[9]*pow(nl,0)-coeff[10]*pow(nl,1)-coeff[11]*pow(nl,2)-coeff[12]*pow(nl,3)

f, ax = plt.subplots()
ax.plot(x,y, color='lightblue', label = 'O3 Column')
ax.plot(x,yfit_svd, label = 'SVD Fit (Fourier+Poly)')
ax.legend(fontsize=12, loc='upper left')
```
